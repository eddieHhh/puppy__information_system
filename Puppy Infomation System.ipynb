# Puppy Infomation System

```python
import os
filename = "puppy.txt"
```

```python
def main():
    while True:
        menu()
        #Ask the user to input an operation
        action = input('Please select the operation you would like to perform\n')
        if action == "1":
            insert() #Insert Puppy Information
        elif action == "2":
            search() #Search Puppy Information
        elif action == "3":
            delete() #Delete Puppy Information 
        elif action == "4":
            modify() #Modify Puppy Information
        elif action == "5":
            sort() #Sort Puppy Information
        elif action == "6":
            total() #Total Number of Puppies
        elif action == "7":
            show() #Show All Puppies Information
        elif action == "0":
            answer = input("Are you sure to exit system? Y/N ")
            if answer == 'Y' or answer == 'y':
                print("Thank you for using Puppy Adoption System!")
                break
            else:
                continue
        else:
            print("Invalid Enter, Please Select Again.")
```

```python
def menu():
    print("=====================Puppy Information Management System=====================")
    print("--------------------------------Function Menu--------------------------------")
    print("\t\t\t1. Insert Puppy Information")
    print("\t\t\t2. Search Puppy Information")
    print("\t\t\t3. Delete Puppy Information")
    print("\t\t\t4. Modify Puppy Information")
    print("\t\t\t5. Sort Puppy Information")
    print("\t\t\t6. Total Number of Puppies")
    print("\t\t\t7. Show All Puppies Information")
    print("\t\t\t0. Exit system")
    print("-----------------------------------------------------------------------------")
```

```python
def save(lst):
    try:
        puppy_txt = open(filename,'a')
        
    except:
        puppy_txt = open(filename,'w')
    for item in lst:
        puppy_txt.write(str(item) + "\n")
    puppy_txt.close()    
```

```python
def insert():
    puppy_list = []
    while True:
        id = input("Please Enter ID: ")
        if not id:
            break
        name = input("Please Enter Name: ")
        if not name:
            break  
        
        try:
            breed = input("Please Enter Breed: ")
            age = int(input("Please Enter Age: "))
            sex = input("Please Enter Sex: ")
            weight = int(input("Please Enter Weight: "))
            height = int(input("Please Enter Height: "))
            color = input("Please Enter Color: ")
            
        except:
            print("Invalid Enter, Please Enter Again.")
            continue
            
        #Save the entered puppy information into the dictionary
        puppy = {"id":id, "name":name, "breed":breed,"age":age,
                "sex":sex,"weight":weight,"height":height,"color":color}
        
        #Add the puppy information to the linked list
        puppy_list.append(puppy) 
        answer = input("Do You Want To Continue Inserting?Y/N\n")
        if answer == 'Y' or answer == 'y':
            continue
        else:
            break
            
    #Call save() function   
    save(puppy_list)
    print("Information For The Puppy Has Been Successfully Inserted!")
    
```

```python
def show_puppy(lst):
    if len(lst) == 0:
        print("No Puppy Information Found, No Data To Display.")
        return
    # Define the display format.
    format = "{:^10}\t{:^10}\t{:^10}\t{:^10}\t{:^10}\t{:^10}\t{:^10}\t{:^10}"
    print(format.format("ID", "Name", "Breed", "Age", "Sex", "Weight", "Height", "Color"))
    for item in lst:
        print(format.format(item.get("id"), item.get("name"),
                                  item.get("breed"), item.get("age"),
                                  item.get("sex"), item.get("weight"),
                                  item.get("height"), item.get("color")))
```

```python
def search():
    puppy_query = []
    while True:
        id = ""
        name = ""
        if os.path.exists(filename):
            method = input("Enter 1 To Search By ID, Enter 2 To Search By Name: ")
            if method == "1":
                id = input("Please Enter Puppy ID ")
            elif method == "2":
                name = input("Please Enter Puppy Name ")
            else:
                print("Invalid Enter, Please Select Again.")
                search()
            with open(filename, 'r') as rfile:
                puppy = rfile.readlines()
                for item in puppy:
                    d = dict(eval(item))
                    if id != "":
                        if d["id"] == id:
                            puppy_query.append(d)
                    elif name != "":
                        if d["name"] == name:
                            puppy_query.append(d)
            #Show the result
            show_puppy(puppy_query)
            #Clear Linked List
            puppy_query.clear()
            answer = input("Do You Want To Continue Searching?Y/N\n")
            if answer == 'Y' or answer == 'y':
                continue
            else:
                break
        else:
            print("No puppy information has been saved yet.")
            return
            
            
```

```python
def delete():
    while True:
        puppy_id = input("Please Enter The ID Of The Puppy You Wish To Delete: ")
        if puppy_id != '':
            if os.path.exists(filename):
                with open(filename, 'r') as file:
                    puppy_record = file.readlines()
            else:
                puppy_record = []
            mark = False #Mark to delete or not
            if puppy_record:
                with open(filename, 'w') as wfile:
                    d = {} #Create an empty dictionary to convert the string into a structured format.
                    for item in puppy_record:
                        d = dict(eval(item))
                        if d['id'] != puppy_id:
                            wfile.write(str(d)+'/n')
                        else:
                            mark = True
                    
                    if mark:
                        print("The puppy with the ID " + puppy_id + "has been deleted.")
                    else:
                        print("The puppy with the ID " + puppy_id + "could not be found.")
            else:
                print("No Puppy Information Available.")
                break
            show()
            answer = input("Do You Want To Continue Deleting?Y/N\n")
            if answer == 'Y' or answer == 'y':
                continue
            else:
                break 
        else:
            print("No Puppy Information Has Been Saved Yet.")  
            return
```

```python
def modify():
    show()
    if os.path.exists(filename):
        with open(filename,'r') as rfile:
            puppy_record = rfile.readlines()
    else:
        return
    puppy_id = input("Please Enter ID: ")
    with open(filename, 'w') as wfile:
        for item in puppy_record:
            d = dict(eval(item))
            if d["id"] == puppy_id:
                print("The Puppy Information Has Been Found And Can Now Be Modified.")
                while True:
                    try:
                        d["breed"] = input("Please Enter Breed: ")
                        d["age"] = int(input("Please Enter Age: "))
                        d["sex"] = input("Please Enter Sex: ")
                        d["weight"] = int(input("Please Enter Weight: "))
                        d["height"] = int(input("Please Enter Height: "))
                        d["color"] = input("Please Enter Color: ")          
                    except:
                        print("Invalid Enter, Please Enter Again.")
                    else:
                        break
                wfile.write(str(d)+'\n')
                print("The Modification Was Successful!")
            else:
                wfile.write(str(d)+'\n')
        answer = input("Do You Want To Continue Modifying?Y/N\n")
        if answer == 'Y' or answer == 'y':
            modify()
        
```

```python
def sort():
    show()
    if os.path.exists(filename):
        with open(filename, 'r') as rfile:
            puppy_lst = rfile.readlines()
        puppy = []
        for item in puppy_lst:
            d = dict(eval(item))
            puppy.append(d)
    else:
        return
    
    asc_or_desc =  input("Enter 1 For Ascending Order, Enter 2 For Descending Order.")
    if asc_or_desc == "1":
        select = False
    elif asc_or_desc == "2":
        select = True
    else:
        print("Invalid Enter, Please Select Again.")
        sort()
    method = input("Please Choose A Sorting Method: Enter 1 To Sort By Age, Enter 2 To Sort By Weight, Enter 3 To Sort By Height: ")
    if method == "1":
        puppy.sort(key = lambda puppy : int(puppy["age"]), reverse = select)
    elif method == "2":
        puppy.sort(key = lambda puppy : int(puppy["weight"]), reverse = select)
    elif method == "3":
        puppy.sort(key = lambda puppy : int(puppy["height"]), reverse = select)
    else:
        print("Invalid Enter, Please Select Again.")
        sort()
    show_puppy(puppy)
```

```python
def total():
    if os.path.exists(filename):
        with open(filename,'r') as rfile:
            puppy = rfile.readlines()
            if puppy:
                print(f"There Are A Total Of {len(puppy)} Puppies.")
            else:
                print("No Puppy Information Available.")
    else:
            print("No Puppy Information Has Been Saved Yet.")
```

```python
def show():
    puppy_lst = []
    if os.path.exists(filename):
        with open(filename,'r') as rfile:
            puppy = rfile.readlines()
            for item in puppy:
                puppy_lst.append(eval(item))
            if puppy:
                show_puppy(puppy_lst)
    else:
        print("No Puppy Information Available.")
```
